```python
from nltk import word_tokenize
text = "I am an ml engineer engineer"
token = word_tokenize(text)
print(token)
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
encoded_text = encode.fit_transform(token)
print(encoded_text)
```
