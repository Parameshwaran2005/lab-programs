```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("training_data.csv")
```

```python
attributes = df.columns[:-1]
target = df.columns[-1]
```

```python
hypothesis = [None for _ in attributes]
```

```python
for index,row in df.iterrows():
    if row[target] == "Yes":
        for i,value in enumerate(row[attributes]):
            if hypothesis[i]==None or hypothesis[i]==value:
                hypothesis[i]= value
            else:
                hypothesis[i] = '?'
                
print(hypothesis)
        
```
