```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
iris = load_iris(as_frame=True)
iris = iris.frame
```

```python
target = iris['target']
```

```python
iris.drop(columns=['target'],inplace=True)
```

```python
pca = PCA(n_components=2)
data_red = pca.fit_transform(iris)
```

```python
red_df = pd.DataFrame(data_red,columns=['PCA1','PCA2'])
```

```python
red_df['target'] = target
```

```python
sns.scatterplot(
    x = red_df['PCA1'],
    y = red_df['PCA2'],
    hue= red_df['target'],
    palette = ['r','g','b'],
    
)
plt.title("PCA")
plt.grid()
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=2)
iris = load_iris(as_frame=True)
iris = iris.frame
target = iris['target']
mapp = {0:'setosa',1: 'versicolor',2: 'virginica'}
target = target.map(mapp)
iris.drop(columns=['target'],inplace=True)
```

```python
load_iris().target_names
```

```python
data_red = pca.fit_transform(iris)
red_data = pd.DataFrame(data_red,columns=['PCA1','PCA2'])
sns.scatterplot(
    x = red_data['PCA1'],
    y = red_data['PCA2'],
    hue = target,
    palette = ['r','g','b']
)
plt.grid()
plt.legend()
plt.show()
```
