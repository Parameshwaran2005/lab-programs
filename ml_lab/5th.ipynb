```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

x = np.random.rand(100)
train, test = x[:50], x[50:]
labels = np.where(train <= 0.5, "Class1", "Class2")

for k in [1,2,3,4,5,20,30]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(train.reshape(-1,1), labels)
    pred = model.predict(test.reshape(-1,1))

    print(f"\nk = {k}")
    for i in range(len(test)):
        print(f"x{i+51} = value: {test[i]:.4f} is classified as {pred[i]}")

    plt.scatter(train, [0]*50, c=np.where(labels=="Class1","b","r"), label="Train")
    plt.scatter(test[pred=="Class1"], [1]*sum(pred=="Class1"), c="b", marker="x", label="Class1")
    plt.scatter(test[pred=="Class2"], [1]*sum(pred=="Class2"), c="r", marker="x", label="Class2")
    plt.title(f"KNN (k={k})")
    plt.legend()
    plt.grid()
    plt.show()
```
