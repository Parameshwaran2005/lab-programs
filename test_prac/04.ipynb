```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('training_data.csv')
attributes = df.columns[:-1]
class_labels = df.columns[-1]
```

```python
hypothesis = [None for _ in attributes]
```

```python
for index,row in df.iterrows():
    if row[class_labels] == "Yes":
        for i,value in enumerate(row[attributes]):
            if hypothesis[i] == None or hypothesis[i] == value:
                hypothesis[i] = value
            else:
                hypothesis[i] = '?'
print(df)
print(f"The final Hypothesis is {hypothesis}")
```
