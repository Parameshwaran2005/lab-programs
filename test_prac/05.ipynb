```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
```

```python
data = np.random.rand(100)
```

```python
train_x,test_x = data[:50],data[50:]
```

```python
train_y = ["Class1" if x<=0.5 else "Class2" for x in train_x]
```

```python
def knn(test_point,k):
    dist = [abs(test_point - x) for x in train_x]
    sorted_labels = [label for _,label in sorted(zip(dist,train_y))]
    return Counter(sorted_labels[:k]).most_common(1)[0][0]
```

```python
dist = [abs(0.45 - x) for x in train_x]
```

```python
for k in [1,2,4]:
    print(f"when k = {k}")
    y_pred = [knn(test_point,k) for test_point in test_x]
    for i,point in enumerate(test_x):
        print(f"point x{51 + i}: (value: {test_x[i]}) is classified as {y_pred[i]}")
    
    plt.scatter(train_x,[0]*len(train_x),
                c = ["blue" if y=="Class1" else "red" for y in train_y],label="Training data")
    plt.scatter(test_x,[1]*len(test_x),
                c = ["blue" if y=="Class1" else "red" for y in y_pred],label="Testing data",marker='x')
    
    plt.legend()
    plt.grid()
    plt.show()
```
