##  Develop a program to implement Principal Component Analysis (PCA) for reducing the dimensionality of the Iris dataset from 4 features to 2.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
data = load_iris(as_frame=True)
data = data.frame
```

```python
data.shape
```

```python
data.head()
```

```python
data.target.value_counts()
```

```python
target = data['target']
data.drop(columns=['target'],inplace=True)
```

```python
pca = PCA(n_components=2)
```

```python
data_reduced = pca.fit_transform(data)
```

```python
reduced_df = pd.DataFrame(data=data_reduced,columns=['PCA1','PCA2'])
```

```python
import seaborn as sns

sns.scatterplot(
    x=data_reduced[:, 0],
    y=data_reduced[:, 1],
    hue=target,
    palette=['r','g','b']
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA on Iris")
plt.grid()
plt.show()
```
