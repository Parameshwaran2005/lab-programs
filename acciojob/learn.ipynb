```python
s1 = set(s1)
s2 = set(s2)

if len(s1)==len(s2) and s1==s2:
    print("True") 
```

```python
l1 = set(s1)

l2 = set(s2)

count1 = {}
count2 = {}
for i in range(len(s1)):
    if s1[i] in count1:
        count1[s1[i]] += 1
    else:
        count1[s1[i]] = 1 
for i in range(len(s2)):
    if s2[i] in count2:
        count2[s2[i]] += 1
    else:
        count2[s2[i]] = 1 
        

if len(s1)==len(s2) and l1==l2 and count1==count2:
    print("True") 



print(l1)
print(l2)
print(len(s1))
print(len(s2))
print(count1)
print(count2)
```

# 2 two sum

```python
nums = [2, 11, 7, 15]
target = 9
```

```python
dict = {}
for i,num in enumerate(nums):
    dict[num] = i
```

```python
for i, num in enumerate(nums):
    need = target - num
    if(need in dict.keys() and dict[need]!=i):
        print(i,dict[need])
```
