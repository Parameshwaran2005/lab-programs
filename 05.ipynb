## 5. Develop a program to implement k-Nearest Neighbour algorithm to classify the randomly generated 100 values of x in the range of [0,1]. Perform the following based on dataset generated.
- a) Label the first 50 points {x1,……,x50} as follows: if (xi ≤ 0.5), then xi ∊ Class1, else xi ∊ Class1
- b) Classify the remaining points, x51,……,x100 using KNN. Perform this for k=1,2,3,4,5,20,30

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
data = np.random.rand(100)
```

```python
train_x, test_x = data[:50],data[50:]
```

```python
train_y = ["Class1" if x<=0.5 else "Class2" for x in train_x]
```

```python
def knn(test_point,k):
    dist = [abs(test_point - x) for x in train_x]
    sorted_labels = [label for _,label in sorted(zip(dist,train_y))]
    return Counter(sorted_labels[:k]).most_common(1)[0][0]
```

```python
Counter([6,6,5,6,7,8,89]).most_common(1)[0][0]
```

```python
sorted(zip([12,4,35,436,36],[245,5,435,4365,653]))
```

```python
for k in [1,2,3,4,5,20,30]:
    y_pred = [knn(point,k) for point in test_x]
    print(f"when k = {k}")
    for i in range(len(test_x)):
        print(f"Point x{i+51} (value: {test_x[i]:.4f}) is classified as {y_pred[i]}")
    plt.scatter(train_x,[0]*len(train_x),
            c=["blue" if y=="Class1" else "red" for y in train_y],
            label="Training data",marker='x')
    plt.scatter(test_x,[1]*len(test_x),
                c = ["blue" if y=="Class1" else "red" for y in y_pred],
                label = "Test data")
    plt.title(f"k = {k}")
    plt.legend()
    plt.grid()
    plt.show()
```
